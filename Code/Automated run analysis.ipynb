```python
import pandas as pd
import numpy as np
import pickle
```

```python
def classifier(x):
    if x < -40:
        return 'ea'
    if x < 0:
        return 'ma'
    if x < 40:
        return 'mi'
    return 'ei'
```

```python
path = '..\\Data\\'
random_state = 42
n_features_to_select = 5
n_runs = 100
df_path = path + 'predictions_seed{}_feat{}_runs{}.csv'.format(random_state, n_features_to_select, n_runs)
rfe_path = path + 'rfe_res_seed{}_feat{}_runs{}.pkl'.format(random_state, n_features_to_select, n_runs)
```

```python
results = pd.read_csv(df_path, header=0, sep=';', decimal=',', index_col=0)
```
