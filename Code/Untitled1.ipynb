```python
means_array = data.groupby(['Features','Type']).median()
means_array = means_array[['R^2', 'RMSE', 'r', 'p']].to_numpy()
```

```python
for j in range(4):
    r = rounds[j]
    print('%{} & {} & {} & & {} & {} & & {} & {} & & {} \\\\'.format(
    row_names[j], np.round(means_array[0,j],r), np.round(means_array[1,j],r), np.round(means_array[2,j],r), 
        np.round(means_array[3,j],r), np.round(means_array[4,j],r),
        np.round(means_array[5,j],r), np.round(means_array[6,j],r)))
```
